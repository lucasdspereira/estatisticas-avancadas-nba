# Evolução de arremessos a cada 100 posses(média das equipes)

```python
fgaevolution = df.groupby('season')['fga_per_100_poss'].mean().reset_index().sort_values(by = 'fga_per_100_poss',
                                                                                        ascending = False)
```

```python
fgaevolution.head()
```

```python
# Plot

df_agrupado = df.groupby('season')['fga_per_100_poss'].mean()
df_agrupado.plot()

plt.title('Evolução de arremessos a cada 100 posses (média das equipes, desde 1974)')
plt.xlabel('Ano')
plt.ylabel('Arremessos a cada 100 posses')

plt.show()
```

No geral, a linha do gráfico tende a acompanhar as tendências de mudanças de regras defensivas da liga. Na década de 1980, as regras eram mais permissivas quanto ao contato físico, sendo este permitido por todos os espaços da quadra. Na década de 1990, a permissão de contato na linha dos 3 pontos foi abolido, o que culminou em mais tentativas de arremessos de quadra. Em 2002, a NBA permitiu as defesas em zona, o que nos dois anos seguintes, resultou em uma queda brusca de tentativas de arremessos, junto a placares mais baixos, gerando críticas dos fãs, que alegavam o jogo estar ficando "chato". Em virtude disso, em 2004, a NBA estendeu a proibição de contatos nas marcações para próximo do aro, o que, desde então, resultou no aumento dos arremessos de quadra. De maneira geral, a linha do gráfico acompanhou as mudanças de dinâmica do jogo ocasionadas pelas mudanças de regras defensivas.

# Mostrando a relação entre Net Rating, aproveitamento de arremessos dos adversários e classificação aos playoffs (1974-2022)

```python
# Selecionamos as colunas dos dataframes de estatísticas avançadas ofensivas e defensivas

colunas_df2 = df2[['team','season','pts_per_100_poss']]
colunas_df4 = df4[['playoffs','opp_pts_per_100_poss', 'opp_fg_percent']]
```

```python
# Em ambas as bases de dados, as colunas 'team' são iguais, o que favoreceu a concatenação

df5 = pd.concat([colunas_df2, colunas_df4], axis=1)
df5.head()
```

```python
df5.tail()
```

```python
df5 = df5[['team', 'season', 'playoffs', 'opp_fg_percent', 'pts_per_100_poss', 'opp_pts_per_100_poss']]
```

```python
df5.head()
```

```python
df5.shape
```

Inserindo Net Rating ( Eficiência Ofensiva - Eficiência Defensiva)

```python
net_rtg = df5['pts_per_100_poss'] - df5['opp_pts_per_100_poss']
```

```python
df5.insert(loc=6, column='net_rtg', value=net_rtg)
```

```python
df5.head()
```

```python
# Plot

sns.set(style = "ticks")

sns.lmplot(x = 'opp_fg_percent',
           y = 'net_rtg',
           data = df5,
           hue = 'playoffs',
           palette = ['tab:orange', 'tab:blue'],
           height = 7)

plt.xlabel('Percentual de arremessos convertidos pelo adversário')
plt.ylabel('Net Rating (Eficiência Ofensiva - Eficiência Defensiva)')
plt.title('Relação entre Net Rating e percentual de arremessos convertidos pelo adversário, entre equipes classificadas e desclassificadas aos playoffs (1974-2022)')

plt.show()
```

Para vencer um jogo de basquete, é necessário fazer mais pontos que o adversário. De maneira geral, times que cedem mais pontos do que fazem, a cada 100 posses, dificilmente vão aos playoffs e, no período considerado da análise, todos os times com Net Rating superior a 5 pontos, o que é um indicativo de bom equilíbrio entre ataque e defesa, se classificaram aos playoffs.
